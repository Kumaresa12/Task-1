**Step 1: Importing the libraries**

```python
import pandas as pd
```

```python
import numpy as np
     
```

**Step 2: Importing dataset**

```python
data = pd.read_csv("/content/Data.csv")
data
```

**Step 3: Handling the missing data**

```python
data.isnull().sum()
```

```python
Sal = (data["Salary"].mean())
Sal
```

```python
data["Salary"] = data["Salary"].fillna(Sal)
data
```

```python
DAge = data["Age"].mean()
DAge  
```

```python
data["Age"] = data["Age"].fillna(DAge)
data
```

**Step 4: Encoding categorical data**

```python
data.Purchased = data.Purchased.map({'Yes':1,'No':0})
data
```

```python
x = data.drop("Purchased",axis=1)
y = data.Purchased
```

**Step 5: Creating a dummy variable**

```python
data_with_dummies = pd.get_dummies(x, columns = ['Country'])
data_with_dummies
```

**Step 6: Splitting the datasets into training sets and Test sets**

```python
import sklearn
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(data_with_dummies,y,test_size=0.3,random_state=42)
```

**Step 7: Feature Scaling**

```python
from sklearn.preprocessing import MinMaxScaler
s=MinMaxScaler(feature_range=(0,1))
```

```python
s.fit(x_train)
scaled_train=s.transform(x_train)
print(scaled_train)
```

```python
s.fit(x_test)
scaled_test=s.transform(x_test)
print(scaled_test)
```
